### 2.1. What is a GeoTIFF file?

* TIFF: Tagged Image File Format for storing raster graphics images.
* GeoTIFF: TIFF file with georeferencing information embedded.

Think of a GeoTIFF file as a collection of actual rasters (we will later deal with them as `numpy.ndarray`) and a collection of metadata that contains georeferencing information. The potential metadata information includes map projection, coordinate systems, and everything else necessary to establish the exact spatial reference for the file. 

*Introduction of the dataset, taken from Lab 5: This file contains a field describing the estimated distribution of all people on Earth in the year 2020. The data comes from the [Gridded Population of the World data set](https://sedac.ciesin.columbia.edu/data/set/gpw-v4-population-count-rev11) produced by a team at Columbia University. We'll be using the version provided at a 15 arcminute or a 0.25 degree resolution (about 30km). This version of the data set describes the total number of individuals that are estimated to have lived in each grid cell in the year 2020.*

Let's check the metadata (data that provides information about other data or data about data) of GeoTIFF file first.

### 2.2. Metadata in `rasterio`

```python
dataset = rasterio.open('Lab5_GPW.tif')
```

We want to know how many layers (bands) there are in the file. A band is like a layer, or a channel. An RGB image, for example, has three channels/bands/layers. Each band shares the same georeferencing information with each other.

```python
dataset.count
```

Or alternatively,

```python
dataset.indexes
```

If you see `(1,)` that means that there is only one band in the raster. `rasterio` starts counting from 1.

We can view the coordinate reference system (CRS) by

```python
dataset.crs
```

The output is EPSG 4326/WGS84: World Geodetic System 1984, commonly used in GPS. This maps a point on earth to a set of longitude, latitude coordinate (notice that epsg 4326 uses the order [lon, lat]). You will try to reproject a geospatial data onto another CRS in Lab 5. 

Next, we can check how many grid cells there are by

```python
# Notice the sequence of height and width here!
# This is a rasterio convention, which can be different from other packages
dataset.shape
```

```python
dataset.height, dataset.width
```

See the bounds of the GeoTIFF 

```python
dataset.bounds
```

```python
# you can of course also view all the metadata
dataset.meta
```

`'nodata'` refers to a special value that rasterio uses to store NaN values. `'transform'` refers to six values that rasterio uses to record the scale and the position of the raster. This can be used to convert x, y coordinate (coord in the specified crs, e.g., in lon, lat) to i, j coordinate (row and col number on a raster, coord in image space).

```python
# This converts x, y to row, col
dataset.index(0, 0)
```

#### Plot the 'numpy array data' using `imshow`

```python
# Define your color palette to use in your heat map
nodes = [0, 0.5, 1]  # positions for each color from 0-1: 0 to vmin, 1 to vmax
color_scheme = ['white', 'yellow', 'red']  # corresponds to nodes
custom_cmap = LinearSegmentedColormap.from_list(
    'custom_name', list(zip(nodes, color_scheme)))
custom_cmap.set_under('gray')  # set values under vmin to gray
custom_cmap.set_over('red')  # set values over vmax to black
```

### 4. Simplify

Plotting shape files might take some time. `simplify` can speed up the plotting procedure as it literally simplifies the coordinate information.

<img src = "simplify.png">

#### Timer functions: tic(), toc()

[Function Source Link](https://stackoverflow.com/questions/5849800/what-is-the-python-equivalent-of-matlabs-tic-and-toc-functions)

```python
import time

def TicTocGenerator():
    # Generator that returns time differences
    ti = 0           # initial time
    tf = time.time() # final time
    while True:
        ti = tf
        tf = time.time()
        yield tf-ti # returns the time difference

TicToc = TicTocGenerator() # create an instance of the TicTocGen generator

# This will be the main function through which we define both tic() and toc()
def toc(tempBool=True):
    # Prints the time difference yielded by generator instance TicToc
    tempTimeInterval = next(TicToc)
    if tempBool:
        print( "Elapsed time: %f seconds.\n" %tempTimeInterval )

def tic():
    # Records a time in TicToc, marks the beginning of a time interval
    toc(False)
```

```python
# Without simplify
tic()
shp = gpd.read_file('zip://MEX_adm0.zip')
shp = shp.geometry[0]

fig, ax = plt.subplots(figsize=(10, 4))
im = ax.imshow(band,
               cmap=custom_cmap,
               vmin=0, vmax=50000,
               extent=(-180, 180, -90, 90))
fig.colorbar(im)
for poly in shp:
    lon, lat = poly.exterior.xy
    ax.plot(lon, lat, 'k-')

ax.set_xlim((shp.bounds[0], shp.bounds[2]))
ax.set_ylim((shp.bounds[1], shp.bounds[3]))

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Population in Mexico (Source: CIESIN 2018)')
plt.show()
toc()
```

```python
# With simplify
tic()
shp = gpd.read_file('zip://MEX_adm0.zip')
shp = shp.geometry[0].simplify(0.1, preserve_topology=False)

fig, ax = plt.subplots(figsize=(10, 4))
im = ax.imshow(band,
               cmap=custom_cmap,
               vmin=0, vmax=50000,
               extent=(-180, 180, -90, 90))
fig.colorbar(im)
for poly in shp:
    lon, lat = poly.exterior.xy
    ax.plot(lon, lat, 'k-')

ax.set_xlim((shp.bounds[0], shp.bounds[2]))
ax.set_ylim((shp.bounds[1], shp.bounds[3]))

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Population in Mexico (Source: CIESIN 2018)')
plt.show()
toc()
```

```python
# With simplify tolerance = 0.16
tic()
shp = gpd.read_file('zip://MEX_adm0.zip')
shp = shp.geometry[0].simplify(0.16, preserve_topology=False)

fig, ax = plt.subplots(figsize=(10, 4))
im = ax.imshow(band,
               cmap=custom_cmap,
               vmin=0, vmax=50000,
               extent=(-180, 180, -90, 90))
fig.colorbar(im)
for poly in shp:
    lon, lat = poly.exterior.xy
    ax.plot(lon, lat, 'k-')

ax.set_xlim((shp.bounds[0], shp.bounds[2]))
ax.set_ylim((shp.bounds[1], shp.bounds[3]))

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Population in Mexico (Source: CIESIN 2018)')
plt.show()
toc()
```

### 5. Masking

How can we remove data which are outside of Mexico? With the `rasterio.mask.mask()` function, we can clip the raster with the shapefile.

```python
clipped_array, _ = rasterio.mask.mask(dataset, shp, nodata = -1)
print(clipped_array.shape)
```

```python
# Remove single-dimensional entries from the shape of an array.
clipped_array = clipped_array.squeeze(axis = 0)
print(clipped_array.shape)
clipped_array
```

```python
# Plotting starts
fig, ax = plt.subplots(figsize=(10, 4))
im = ax.imshow(clipped_array, cmap=custom_cmap,
               extent=(-180, 180, -90, 90),
               vmin=0, vmax=50000)
fig.colorbar(im)

# Zoom in
ax.set_xlim((shp.bounds[0], shp.bounds[2]))
ax.set_ylim((shp.bounds[1], shp.bounds[3]))

# Label axes and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Population in Mexico (Source: CIESIN 2018)')
plt.show()
```
