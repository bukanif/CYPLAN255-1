```python
from matplotlib import pyplot as plt
import pandas as pd
colors = ['b','g','r','c','m','y','orange','purple']
```

```python
sv = pd.read_csv('data/Remittances Survey.csv')
sv = sv.rename(columns={ 'Are you an immigrant / member of a diaspora community?':'immigrant',
                          'If so, what country are you from?':'country_from',
                        'Do you send money (or goods) to loved ones in another country?':'send_rem',
                        'What country do you send money/goods to? (Select one)':'send_country',
                       'Who are the recipients of what you send?':'recipients',
                       'What platform(s) do you use to send these things to your loved ones?':'plat_used',
                       'What do your loved ones spend the cash remittances on?':'spending',
                       'As a resource sender, would you/your family consider directly paying for clean energy products, rather than sending cash?':'demand',
                    '(Optional) Are there any factors making it harder for you/your loved ones to SEND or RECEIVE the remittances? (ie. fees, wait time, etc.)':'barriers' 
                       })
sv
```

```python
plt.figure(figsize=(10,7))
plt.title('What country do you send money/goods to?', fontsize=16)
sv.value_counts('send_country').plot(kind='bar', color=['c','g','r','r','r','r','m','m','m','m','m','m','m','m','m','m','m','m','m','m','m','m'])
```

```python
plt.figure(figsize=(10,7))
sv.value_counts('send_rem').plot(kind='pie', colors=colors, fontsize=14, ylabel="")
plt.title('Do you send money (or goods) to loved ones in another country?',fontsize=14)
plt.legend(labels=sv.value_counts('send_rem'), loc='upper right')
plt.show()
```

```python
recip = {'Parents':16 , 'Children':1, 'Siblings':9,'Grandparents':6,'Aunts, uncles, cousins':9,'Greater community':6, 'Friends':3} 
recip = pd.Series(recip).sort_values(ascending=False)
plt.figure(figsize=(10,7))
recip.plot(kind='bar', color=colors, fontsize=12)
plt.title('Who are the recipients of what you send?',fontsize=14)
```

```python
spend = {'Household':27 ,'Education':10,'Business':2,'Medical':10,'Saving':3, 'IDK':2 } 
spend = pd.Series(spend).sort_values(ascending=False)
fig,ax = plt.subplots(figsize=(10,7))
plt.title('What expenses do your loved ones spend the cash remittances on?', fontsize=16)
spend.plot(kind='bar',color=colors)
```

```python
plat = {'Western Union':19,'Moneygram':5,'Bank of China':3,'Xoom':2,'World Remit':1,'Mpesa':1,'Indian Bank':1,'Remitly': 1,'Chipper':1,'Binance':1, 'Sendwave':1, 'Zelle':1} 
plat = pd.Series(plat)
plt.figure(figsize=(10,7))
plat.plot(kind='bar', color=colors)
plt.title('What platform(s) do you use to send money to your loved ones?',fontsize=14)
```

```python
plt.figure(figsize=(10,7))
sv.value_counts('demand').plot(kind='pie', colors=['r','orange','g'], ylabel='')
plt.legend(labels=sv.value_counts('demand'), loc='upper right')
plt.title('Would you consider directly paying for clean energy products, rather than sending cash?',fontsize=14)
plt.show()
```
